```python
train = train.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'MasVnrType'], axis=1)
test = test.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'MasVnrType'], axis=1)
```

```python
print(train.shape)
print(test.shape)
```

```python
train = train.dropna(subset=['GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond'], axis=0)
test = test.dropna(subset=['GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond'], axis=0)
```

```python
train_copy = train.copy()
test_copy = test.copy()
```

```python
train_copy[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']] = train_copy[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']].fillna('No')
```

```python
test_copy[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']] = test_copy[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']].fillna('No')
```

```python
# Train
category_mode = train_copy['Electrical'].mode()[0]  # Mode
train_copy['Electrical'] = train_copy['Electrical'].fillna(category_mode)

# Mengisi missing values untuk data numerik dengan median
numeric_median = train_copy['LotFrontage'].median()  # Median
train_copy['LotFrontage'] = train_copy['LotFrontage'].fillna(numeric_median)
```

```python
numeric_median_2 = train_copy['MasVnrArea'].median()  # Median
train_copy['MasVnrArea'] = train_copy['MasVnrArea'].fillna(numeric_median_2)
```

```python
# Test
category_mode_test = test_copy[['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'SaleType']].mode().iloc[0]  # Mode
test_copy[['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'SaleType']] = test_copy[['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'SaleType']].fillna(category_mode_test)

# Mengisi missing values untuk data numerik dengan median
numeric_median_test = test_copy[['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']].median()  # Median
test_copy[['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']] = test_copy[['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']].fillna(numeric_median_test)
```

```python
numeric_median_test_2 = test_copy[['LotFrontage', 'MasVnrArea']].median()  # Median
test_copy[['LotFrontage', 'MasVnrArea']] = test_copy[['LotFrontage', 'MasVnrArea']].fillna(numeric_median_test_2)
```

## Feature Engineering

```python
# 1. Menjumlahkan semua luas fasilitas yang ada sehingga menghasilkan luas rumah dalam square feet
train_copy['TotalSquareFeet'] = train_copy['1stFlrSF'] + train_copy['2ndFlrSF'] + train_copy['GrLivArea'] + train_copy['GarageArea'] + train_copy['TotalBsmtSF']
test_copy['TotalSquareFeet'] = test_copy['1stFlrSF'] + test_copy['2ndFlrSF'] + test_copy['GrLivArea'] + test_copy['GarageArea'] + test_copy['TotalBsmtSF']
```

```python
# 2. Menjumlahkan tiap-tiap kamar mandi sehingga mendapatkan total kamar mandi yang ada
train_copy['TotalBath'] = train_copy['BsmtFullBath'] + train_copy['BsmtHalfBath'] + train_copy['FullBath'] + train_copy['HalfBath']
test_copy['TotalBath'] = test_copy['BsmtFullBath'] + test_copy['BsmtHalfBath'] + test_copy['FullBath'] + test_copy['HalfBath']
```

```python
# 6. Mengalikan Luas garasi dan mobil yang dapat ditampung
train_copy['GarageScore'] = train_copy['GarageArea'] * train_copy['GarageCars']
test_copy['GarageScore'] = test_copy['GarageArea'] * test_copy['GarageCars']
```

```python
# 7. Mengalikan kualitas dan kondisi keseluruhan bangunan
train_copy['OverallQualCond'] = train_copy['OverallQual'] * train_copy['OverallCond']
test_copy['OverallQualCond'] = test_copy['OverallQual'] * test_copy['OverallCond']
```

```python
from pycaret.regression import setup, create_model, compare_models, predict_model
```

```python
regression = setup(train_copy, target='SalePrice', categorical_features=['MSZoning', 'Street', 'LotShape',
                                                                         'LandContour', 'Utilities', 'LotConfig',
                                                                         'LandSlope', 'Neighborhood', 'Condition1',
                                                                         'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle',
                                                                         'RoofMatl', 'Exterior1st', 'Exterior2nd', 'ExterQual',
                                                                         'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
                                                                         'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
                                                                         'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
                                                                         'Functional', 'GarageType', 'GarageFinish', 'GarageQual',
                                                                         'GarageCond', 'PavedDrive', 'SaleType', 'SaleCondition',],
                   preprocess=True, normalize=True, normalize_method='minmax',
                   remove_outliers=True, outliers_threshold=0.05, session_id=123, numeric_imputation='median')
```

```python
lr = create_model('lr')
```

```python
best_model = compare_models()

# Menampilkan ringkasan hasil
print(best_model)
```

```python
# Prediksi pada test set
predictions = predict_model(best_model)

# Menampilkan metrik performa
print(predictions)
```
